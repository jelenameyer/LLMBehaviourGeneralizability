```python
#read csvs

import pandas as pd

dataNoTemplate = pd.read_csv("dospert_llm_scores_all_candidates.csv")
dataWithTemplate = pd.read_csv("dospert_llm_scores_chatstyle.csv")
dataWithTemplate
```

```python
# Get the column (1-5) with the largest value in each row
dataWithTemplate['max_column'] = dataWithTemplate[["1", "2", "3", "4", "5"]].idxmax(axis=1)
dataNoTemplate['max_column'] = dataNoTemplate[["1", "2", "3", "4", "5"]].idxmax(axis=1)

dataNoTemplate
```

```python
df1 = dataWithTemplate[['max_column']].rename(columns={'max_column': 'max_column_template'})
df2 = dataNoTemplate[['max_column']].rename(columns={'max_column': 'max_column_no_template'})

# Concatenate them side by side
merged_df = pd.concat([df1, df2], axis=1)
merged_df["same"] = merged_df["max_column_template"] == merged_df["max_column_no_template"]
merged_df["same"].sum()
```
