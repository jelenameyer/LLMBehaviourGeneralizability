# Analyse Variance and differences between no chat and chat in BART Task

```python
import pandas as pd

BART_WITH_CHAT = pd.read_csv("outputs/SmolLM2-1.7B-Instruct_Bart_Prompting_results.csv")
BART_NO_CHAT = pd.read_csv("outputs/SmolLM2-1.7B-Instruct_Bart_Prompting_noChatTemplate_results.csv")
```

```python
summary_stats_chat = BART_WITH_CHAT[["log_prob_pump", "log_prob_stop"]].describe()
summary_stats_no_chat = BART_NO_CHAT[["log_prob_pump", "log_prob_stop"]].describe()
summary_stats_df = pd.concat([summary_stats_chat, summary_stats_no_chat], axis=1)
summary_stats_df
```

```python
# Get the column with the largest value in each row
BART_WITH_CHAT['max_column'] = BART_WITH_CHAT[["log_prob_pump", "log_prob_stop"]].idxmax(axis=1)
BART_NO_CHAT['max_column'] = BART_NO_CHAT[["log_prob_pump", "log_prob_stop"]].idxmax(axis=1)
```

```python
df1 = BART_WITH_CHAT[['max_column']].rename(columns={'max_column': 'max_column_template'})
df2 = BART_NO_CHAT[['max_column']].rename(columns={'max_column': 'max_column_no_template'})

# Concatenate them side by side
merged_df = pd.concat([df1, df2], axis=1)
merged_df["same"] = merged_df["max_column_template"] == merged_df["max_column_no_template"]
same = merged_df["same"].sum()
complete_data = len(merged_df)

(same * 100)/complete_data
```
